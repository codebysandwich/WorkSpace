```python
import pandas as pd
from datetime import timedelta
```

```python
file = r'./docfiles/外出打卡-杭雪琴-2021-11-17(1).xlsx'
time_cols = ['创建日期', '时间']
df = pd.read_excel(file, parse_dates=time_cols)
df['日期'] = df['时间'].dt.date
df.head(10)
```

```python
groups = df.groupby(['工号', '日期'])
```

```python
def cal(df):
    count = df['时间'].count()
    time_delta = round((df['时间'].max() - df['时间'].min()).seconds / 3600, 2)
    count1 = ((df['时间'].dt.hour >= 19) & (df['时间'].dt.hour < 21)).sum()
    count2 = (df['时间'].dt.hour >= 21).sum()
    return pd.Series({'姓名': df['姓名'].iloc[0], '打卡次数': count, '打卡时间差': time_delta, '19-21点签到次数': count1, 
                      '21点后签到次数': count2})
```

```python
res = groups.apply(cal)  #.to_excel('打卡情况明细.xlsx')
```

```python
p1 = res[res['打卡时间差'] >= 6].groupby(['工号', '姓名'])[['打卡次数']].count()
```

```python
p2 = res[res['打卡时间差'] >= 6].groupby(['工号', '姓名'])[['19-21点签到次数', '21点后签到次数']].sum()
```

```python
pd.concat([p1, p2], axis=1)
```
